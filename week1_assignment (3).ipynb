#### Task 7: Sum of Numbers
1. Write a program that uses a while loop to calculate the sum of all numbers from 1 to a number entered by the user.

````python
# This program calculates the sum of all numbers from 1 to a number entered by the user

number = int (input("Enter a positive integer: "))
total_sum = 0
counter = 1
while counter <= number:
    total_sum += counter
    counter += 1
print(f"The sum of all numbers from 1 to {number} is {total_sum}")
    
````
